```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import yfinance as yf
```

```python
# Load the data from a CSV file
# Define the list of stocks you want to download data for
stocks = ['RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'INFY.NS', 'HINDUNILVR.NS']

# Download the data using yfinance
data = yf.download(stocks, start='2015-01-01', end='2023-03-14', group_by='ticker')
```

```python
data_new = data[[(       'TCS.NS',     'Close'),('HINDUNILVR.NS',     'Close'),(      'INFY.NS',     'Close'),(  'HDFCBANK.NS',     'Close'),(  'RELIANCE.NS',     'Close')]]
```

```python
# Calculate the daily returns
returns = data_new.pct_change().dropna()
```

```python
# Define the prior distribution
mu_prior = np.mean(returns)
std_prior = np.std(returns)
prior = norm(mu_prior, std_prior)
```

```python
# Calculate the likelihood
def likelihood(mu, std):
    return np.prod(norm(mu, std).pdf(returns))
# Define the posterior distribution
def posterior(mu, std):
    return likelihood(mu, std) * prior.pdf(mu)
# Find the MAP estimates
mu_map, std_map = max([(mu, std) for mu in np.linspace(mu_prior-0.05, mu_prior+0.05, 100) for std in np.linspace(std_prior-0.05, std_prior+0.05, 100)], key=lambda x: posterior(*x))

```

```python
# Print the results
print(f'MAP estimate of mean: {mu_map:.4f}')
print(f'MAP estimate of standard deviation: {std_map:.4f}')
```
